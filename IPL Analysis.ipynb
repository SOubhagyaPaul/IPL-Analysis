```python
# find out v kholi scored max in against teams
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
def run_scored(bats_name):
    vk=delivery[delivery['batsman']==bats_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
run_scored('V Kohli')
```

```python
#isin
#find the most destructive death over batsman in the history of IPL
# Strick rate=(Number of runs/number of balls)/100
#min batsman 200 ball in over 16-20

death_over=delivery[delivery['over']>15]
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
#runs scored by this batsman and also the the number of ball they played

final=delivery[delivery['batsman'].isin(batsman_list)]

runs=final.groupby('batsman')['batsman_runs'].sum()
balls=final.groupby('batsman')['batsman_runs'].count()
sr=(runs/balls)*100
sr
```
